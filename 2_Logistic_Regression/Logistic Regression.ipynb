# Logistic Regression

In this part, I will build a logistic regression model to
predict whether a student gets admitted into a university.

```python
%matplotlib inline
from importlib import reload
import scipy.optimize as op
import pandas as pd
from sklearn.metrics import confusion_matrix as cm
```

```python
import module
from module import *
reload(module=module)
```

##  Visualizing the data

```python
data1=np.loadtxt('data1.txt',delimiter=',')
X = data1[:,[0,1]]
y = data1[:,2][:,None]
```

```python
plotData(X,y)
```

## Compute Cost and Gradient

```python
m,n = X.shape
X1 = np.vstack([np.ones(m), X.T]).T

initial_theta = np.zeros([n + 1, 1])
cost, gradient = costFun(initial_theta, X1, y,lam=0),grad(initial_theta,X1,y,lam=0)
```

```python
cost,gradient
```

```python
# Minimising using Newton Conjugate Gradient
lam=0
result = op.minimize(fun=costFun,args=(X1,y,lam),x0=initial_theta,method='TNC',jac=grad)
```

```python
theta = result['x']
result
```

## Plotting decision boundary

```python
plotDecisionBoundary(theta=theta,X=X1,y=y)
plt.title('Training data with decision boundary');
```

```python
y_pred=predict(theta,X1)
acuracy=np.sum(y==y_pred)/len(y)
print('Accuracy:', acuracy)
```

```python
#confusion matrix
df = pd.DataFrame({'actual':y.ravel(),'predicted':y_pred.ravel()})
pd.crosstab(df.actual,df.predicted)
```

```python
cm(y_pred=y_pred,y_true=y)
```

# Regularized logistic regression

In this part, I will implement regularized logistic regression
to predict whether microchips from a fabrication plant passes quality assurance (QA). 
During QA, each microchip goes through various tests to ensure it is functioning correctly.

```python
import module
from module import *
reload(module=module)
```

## Vizualising data

```python
data2 = np.loadtxt('data2.txt',delimiter=',')
X = data2[:,[0,1]]; y = data2[:,2][:,None]
plotData(X,y)
ax = plt.gca()
ax.legend(('y=1','y=0'))
ax.set_xlabel('Microchip Test 1')
ax.set_ylabel('Microchip Test 2');
```

```python
# Generating higher order features
X2 = mapFeature(X[:,0], X[:,1])
X2.shape
```

```python
#Initialize fitting parameters
initial_theta = np.zeros([X2.shape[1], 1])
#Set regularization parameter lambda to 1
lam = 1
# initial cost and gradient for regularized logistic regression

cost, gradient = costFun(initial_theta, X2, y, lam), grad(initial_theta,X2,y,lam)
print(cost,gradient)
```

## Regularization and Accuracies

Trying the following values of lambda (0, 1, 10, 100)  and see how  decision boundary changes

```python
lam=0
result = op.minimize(fun=costFun,args=(X2,y,lam),
                     x0=initial_theta,method='TNC',jac=grad)
theta = result.x
plotDecisionBoundary(theta,X2,y)
ax=plt.gca()
ax.set_title('lambda = {}'.format(lam))
ax.set_xlabel('Microchip Test 1')
ax.set_ylabel('Microchip Test 2')
ax.legend(['y = 1', 'y = 0', 'Decision boundary']);

# Compute accuracy on our training set
y_pred=predict(theta,X2)
acuracy=np.sum(y==y_pred)/len(y)
print('Accuracy:', acuracy)
```

```python
lam=1
result = op.minimize(fun=costFun,args=(X2,y,lam),
                     x0=initial_theta,method='TNC',jac=grad)
theta = result.x
plotDecisionBoundary(theta,X2,y)
ax=plt.gca()
ax.set_title('lambda = {}'.format(lam))
ax.set_xlabel('Microchip Test 1')
ax.set_ylabel('Microchip Test 2')
ax.legend(['y = 1', 'y = 0', 'Decision boundary']);

# Compute accuracy on our training set
y_pred=predict(theta,X2)
acuracy=np.sum(y==y_pred)/len(y)
print('Accuracy:', acuracy)
```

```python
lam=10
result = op.minimize(fun=costFun,args=(X2,y,lam),
                     x0=initial_theta,method='TNC',jac=grad)
theta = result.x
plotDecisionBoundary(theta,X2,y)
ax=plt.gca()
ax.set_title('lambda = {}'.format(lam))
ax.set_xlabel('Microchip Test 1')
ax.set_ylabel('Microchip Test 2')
ax.legend(['y = 1', 'y = 0', 'Decision boundary']);

# Compute accuracy on our training set
y_pred=predict(theta,X2)
acuracy=np.sum(y==y_pred)/len(y)
print('Accuracy:', acuracy)
```
