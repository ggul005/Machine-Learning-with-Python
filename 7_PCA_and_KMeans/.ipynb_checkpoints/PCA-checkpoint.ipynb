##  Load Example Dataset

```python
data1 = loadmat('data1.mat')
X = data1['X']
```

```python
# Visualize the example dataset
plt.plot(X[:, 0], X[:, 1], 'bo')
plt.axis([0.5, 6.5, 2, 8]); plt.axis('square');
```

##  Principal Component Analysis

```python
#  Before running PCA, it is important to first normalize X
X_norm, mu, sigma = featureNormalize(X)
```

```python
#  Run PCA
U, S = pca(X_norm)
```

```python
#  Draw the eigenvectors centered at mean of data. These lines show the
#  directions of maximum variations in the dataset.

plt.plot(X[:, 0], X[:, 1], 'bo')
plt.axis([0.5, 6.5, 2, 8]); plt.axis('square');
drawLine(mu, mu + 1.5 * S[0] * U[:,0].T, ls='-', lw=2,c='k')
drawLine(mu, mu + 1.5 * S[1] * U[:,1].T, ls='-', lw=2,c='k')
```

```python
print('Top eigenvector:')
print(' U[:,0] = {0:.3f}  {1:.3f}'.format(U[1,0], U[0,0]))
```

##  Dimension Reduction

```python
#  Plot the normalized dataset (returned from pca)
plt.plot(X_norm[:, 0], X_norm[:,1], 'bo')
plt.axis([-4, 3, -4, 3]); plt.axis('square');
```

```python
#  Project the data onto K = 1 dimension
K = 1
Z = projectData(X_norm, U, K)
print('Projection of the first example: {0:.3f}'.format(Z[0][0]))
```

```python
X_rec = recoverData(Z,U,K)
```

```python
#  Draw lines connecting the projected points to the original points
plt.plot(X_norm[:, 0], X_norm[:,1], 'bo')
plt.plot(X_rec[:, 0], X_rec[:, 1], 'ro')

for i in range(X_norm.shape[0]):
    drawLine(X_norm[i,:], X_rec[i,:], ls='--', lw= 1,c='k');

plt.axis([-3, 3, -3, 3]); plt.axis('square');
```

##  PCA for Visualization

```python
A = plt.imread('bird_small.png')
n1,n2, _ = A.shape
```

```python
X = A.reshape([n1*n2,3])
K = 16; 
max_iters = 10
initial_centroids = kMeansInitCentroids(X, K)
centroids, idx = runkMeans(X, initial_centroids, max_iters)
idx = idx.astype('int');
```

```python
#  Sample 1000 random indexes (since working with all the data is
#  too expensive. If you have a fast computer, you may increase this.
sel = np.floor(np.random.rand(1000) * X.shape[0]).reshape([-1,1]).astype('int').flatten()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
for k in range(K):
        indx = idx[sel,0] == k
        indx = indx.flatten()
        rgba = plt.cm.hsv(20*k)
        ax.scatter(X[indx,0], X[indx,1],X[indx,2], s=10,c=rgba)
ax.set_title('Pixel dataset plotted in 3D. Color shows centroid memberships');
#ax.scatter(X[sel,0],X[sel,1],X[sel,2],s=10);
```
