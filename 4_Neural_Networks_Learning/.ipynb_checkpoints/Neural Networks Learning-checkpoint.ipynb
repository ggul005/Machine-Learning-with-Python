```python
# Setup the parameters 
input_layer_size  = 400  # 20x20 Input Images of Digits
hidden_layer_size = 25   # 25 hidden units
num_labels = 10
```

##  Loading and Visualizing Data

```python
data1 = loadmat('data1.mat')
X = data1['X']
y = data1['y']
indx = y == 10; y[indx]=0
m=len(y)
```

## Loading Parameters
 We load some pre-initialized neural network parameters.


```python
# Loading Saved Neural Network Pameters 
data2 =  sio.loadmat('weights.mat')
theta1 = data2['Theta1']
theta2 = data2['Theta2']
```

```python
# Unroll parameters 
nn_params = np.hstack([theta1.ravel(), theta2.ravel()])#.reshape([-1,1])
```

## Compute Cost (Feedforward)

```python
lam = 0

J1,grad = nnCostFunction(nn_params, input_layer_size, hidden_layer_size, \
                   num_labels, X, y, lam)

print(J1)
```

## Initializing Pameters


```python
#Initializing Neural Network Parameters
np.random.seed(123)
initial_Theta1 = randInitializeWeights(input_layer_size, hidden_layer_size)
initial_Theta2 = randInitializeWeights(hidden_layer_size, num_labels)

# Unroll parameters
initial_nn_params = np.hstack([initial_Theta1.ravel(), initial_Theta2.ravel()])
```

## Training NN
To train our neural network, we will now use "fmincg", which
is a function which works similarly to "fminunc". These
advanced optimizers are able to train our cost functions efficiently as
long as we provide them with the gradient computations.

```python
lam = 1

# Create "short hand" for the cost function to be minimized and associated grad
costFun = lambda x0,lam0: nnCostFunction(x0,input_layer_size, \
                                   hidden_layer_size, \
                                   num_labels, X, y, lam0)[0]
grad = lambda x0,lam0: nnCostFunction(x0,input_layer_size, \
                                   hidden_layer_size, \
                                   num_labels, X, y, lam0)[1]
```

```python
#X1 = np.hstack([np.ones(m).reshape([-1,1]),X])
%time result = op.fmin_cg(f=costFun,args=(lam,),\
                     x0=initial_nn_params,fprime=grad,disp=1)
```

```python
# Obtain Theta1 and Theta2 back from nn_params
Theta1 = np.reshape(result[:hidden_layer_size * (input_layer_size + 1)], \
                 [hidden_layer_size, (input_layer_size + 1)])

Theta2 = np.reshape(result[hidden_layer_size * (input_layer_size + 1):], \
                 [num_labels, (hidden_layer_size + 1)])
```

##  Implement Predict 
After training the neural network, we would like to use it to predict
the labels. 

```python
pred = predict(Theta1,Theta2,X)

print('Training Set Accuracy: %.3f' % np.mean(pred==y))
```
