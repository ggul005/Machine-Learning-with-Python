##  K-Means Clustering on Pixels 
Here we will use K-Means to compress an image. To do this,
we will first run K-Means on the colors of the pixels in the image and
then we will map each pixel on to it's closest centroid.

```python
#  Load an image of a bird
A = plt.imread('bird_small.png')
```

```python
# Reshape the image into an Nx3 matrix where N = number of pixels.
# Each row will contain the Red, Green and Blue pixel values
# This gives us our dataset matrix X that we will use K-Means on.
n1,n2, _ = A.shape
X = A.reshape([n1*n2,3])
```

```python
# Run the K-Means algorithm on this data
K = 16
max_iters = 10
initial_centroids = kMeansInitCentroids(X, K)
centroids, idx = runkMeans(X, initial_centroids, max_iters)
```

##  Image Compression

```python
#idx_c = findClosestCentroids(X, centroids)
idx.astype('int')
```

```python
A_recovered = centroids[idx.astype('int'),:].reshape([n1,n2,3])
```

```python
fig = plt.figure(figsize=[10,4])
ax1 = fig.add_subplot(121); ax1.set_axis_off()
ax2 = fig.add_subplot(122); ax2.set_axis_off()
ax1.imshow(A); ax1.set_title('Original')
ax2.imshow(A_recovered); ax2.set_title('Compressed, with {0:d} colors.'.format(K));
```
